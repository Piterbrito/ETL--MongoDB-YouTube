```python
# Function to Enter all the Json files


def jsonOpen1(name):

    with open(name, 'r') as handle:
        parsed = json.load(handle)

    name = (json.dumps(parsed, indent=4, sort_keys=True))

    return print(name)



# Calling the function for each json file and saving in variable

youtube_KR = jsonOpen1(youtube_KR)
youtube_MX = jsonOpen1(youtube_MX)
youtube_RU = jsonOpen1(youtube_RU)
```

```python
# Function that cleans and filter the data

def clean1 (df):

    # Geting the meaninful columns
    df = df[["video_id","title", "channel_title","publish_time",
                              "tags","views", "likes", "dislikes", "comment_count"  ]]

    # Cleaning overload data
    df["publish_time"] = df["publish_time"].map(lambda x: str(x)[:-14])


    #Cleaning symbols 
    df['channel_title']= df['channel_title'].map(lambda x: x.lstrip('+-@#$%^&*पीरियड्स के समय, पेट ').rstrip('aAbBcC'))

    return df.head()


youtube_US = clean1(youtube_US)
youtube_CA = clean1(youtube_CA)
youtube_DE = clean1(youtube_DE)
youtube_FR = clean1(youtube_FR)
youtube_GB = clean1(youtube_GB)
youtube_IN = clean1(youtube_IN)
```
