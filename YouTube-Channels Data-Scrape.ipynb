```python
soup = BeautifulSoup(html.text)

body = soup.find_all({'div': {'class': 'table-body'}})
```

```python
def get_table (row):
    
    #picking a table and sending into a dict
    first = [i.text for i in row if i != u'/']
             
    return dict(rank=int(first[0]),
               qrade= str(first[1]),
                channel= str(first[2]),
                videos= float(first[3].replace(",", "")),
                subscribers= float(first[4].replace(",", "") ),
                views= float(first[5].replace(",", ""))
               )
```
